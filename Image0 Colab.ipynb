```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Task 1: Read the provided image “Lena Image”
lena_image = cv2.imread('/content/Lena_Image.png')  # Change the path if needed

# Task 2: Display the Lena image
plt.imshow(cv2.cvtColor(lena_image, cv2.COLOR_BGR2RGB))
plt.title("Lena Image")
plt.axis("off")
plt.show()
```

```python
# Task 3: Matrix Representation of the Lena Image
print("Matrix Representation of Lena Image:")
print(lena_image)
```

```python
# Task 4: Perform Arithmetic Operations on Image_1 and Image_2
image1 = cv2.imread('/content/Image 1.1.jpg')
image2 = cv2.imread('/content/Image 1.2.jpg')

addition = cv2.add(image1, image2)
subtraction = cv2.subtract(image1, image2)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(addition, cv2.COLOR_BGR2RGB))
plt.title("Addition")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(subtraction, cv2.COLOR_BGR2RGB))
plt.title("Subtraction")
plt.axis("off")
plt.show()
```

```python
# Task 5: Perform Bitwise Operations on Image_3 and Image_4
image3 = cv2.imread('/content/Image_1.png', 0)
image4 = cv2.imread('/content/Image_2.png', 0)

bitwise_and = cv2.bitwise_and(image3, image4)
bitwise_or = cv2.bitwise_or(image3, image4)
bitwise_not = cv2.bitwise_not(image3)
bitwise_xor = cv2.bitwise_xor(image3, image4)

plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.imshow(bitwise_and, cmap='gray')
plt.title("Bitwise AND")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(bitwise_or, cmap='gray')
plt.title("Bitwise OR")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(bitwise_not, cmap='gray')
plt.title("Bitwise NOT (Image_3)")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(bitwise_xor, cmap='gray')
plt.title("Bitwise XOR")
plt.axis("off")
plt.show()
```

```python
# Task 6: Convert Colour “Lena” Image to Grayscale Image
gray_lena = cv2.cvtColor(lena_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_lena, cmap='gray')
plt.title("Grayscale Lena Image")
plt.axis("off")
plt.show()

```

```python
# Task 7: Convert Grayscale “Lena” Image to Binary Image
_, binary_lena = cv2.threshold(gray_lena, 127, 255, cv2.THRESH_BINARY)
plt.imshow(binary_lena, cmap='gray')
plt.title("Binary Lena Image")
plt.axis("off")
plt.show()
```

```python
# Task 8: Resize the Lena Image to 256 x 256
resized_lena = cv2.resize(lena_image, (256, 256))
plt.imshow(cv2.cvtColor(resized_lena, cv2.COLOR_BGR2RGB))
plt.title("Resized Lena Image (256x256)")
plt.axis("off")
plt.show()
```
