```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
from tensorflow.keras.datasets import mnist
from sklearn.preprocessing import label_binarize

# Load image
image = cv2.imread('/content/Screenshot 2025-01-30 at 12.24.25â€¯PM.png')
```

```python
# 1.1 Image Resizing
resized_linear = cv2.resize(image, (200, 200), interpolation=cv2.INTER_LINEAR)
resized_nn = cv2.resize(image, (200, 200), interpolation=cv2.INTER_NEAREST)
resized_poly = cv2.resize(image, (200, 200), interpolation=cv2.INTER_CUBIC)

# 1.2 Image Blurring
box_blur = cv2.blur(image, (5, 5))
gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)
adaptive_blur = cv2.medianBlur(image, 5)

# Display Images
fig, axes = plt.subplots(2, 4, figsize=(12, 6))
axes[0, 0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)); axes[0, 0].set_title('Original')
axes[0, 1].imshow(cv2.cvtColor(resized_linear, cv2.COLOR_BGR2RGB)); axes[0, 1].set_title('Linear Resizing')
axes[0, 2].imshow(cv2.cvtColor(resized_nn, cv2.COLOR_BGR2RGB)); axes[0, 2].set_title('Nearest Neighbor')
axes[0, 3].imshow(cv2.cvtColor(resized_poly, cv2.COLOR_BGR2RGB)); axes[0, 3].set_title('Polynomial Resizing')
axes[1, 0].imshow(cv2.cvtColor(box_blur, cv2.COLOR_BGR2RGB)); axes[1, 0].set_title('Box Blur')
axes[1, 1].imshow(cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2RGB)); axes[1, 1].set_title('Gaussian Blur')
axes[1, 2].imshow(cv2.cvtColor(adaptive_blur, cv2.COLOR_BGR2RGB)); axes[1, 2].set_title('Adaptive Blur')

for ax in axes.flat:
    ax.axis('off')
plt.show()
```

```python
# Task-2: Machine Learning on MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess Data
x_train, x_test = x_train.reshape(-1, 784) / 255.0, x_test.reshape(-1, 784) / 255.0
y_train_bin, y_test_bin = label_binarize(y_train, classes=range(10)), label_binarize(y_test, classes=range(10))

# Select Two Algorithms
models = {
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True)
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    print(f"\n{name} Model:")
    scores = cross_val_score(model, x_train, y_train, cv=kf, scoring='accuracy')
    print(f"Mean Accuracy: {np.mean(scores):.4f}")

    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)

    # Performance Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    conf_matrix = confusion_matrix(y_test, y_pred)
    y_prob = model.predict_proba(x_test)
    fpr, tpr, _ = roc_curve(y_test_bin.ravel(), y_prob.ravel())
    roc_auc = auc(fpr, tpr)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Confusion Matrix:\n{conf_matrix}")

    # ROC Curve
    plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f}')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()

plt.show()
```
