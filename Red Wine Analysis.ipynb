```python
import numpy as np
import matplotlib.pyplot as plt
XY=np.genfromtxt("/content/drive/MyDrive/winequality-red.csv",delimiter=";",skip_header=1)
X=XY[:,0:12]
Y=XY[:,-1]

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

# Bar graph
unique_values, counts = np.unique(Y, return_counts=True)
axes[0].bar(unique_values, counts)
axes[0].set_title('Bar Graph of Unique Values in Y')
axes[0].set_xlabel('Values')
axes[0].set_ylabel('Counts')

# Histogram
axes[1].hist(X, bins=20, alpha=0.7, label=['Column 1', 'Column 2', 'Column 3'])
axes[1].set_title('Histograms of Different Columns in X')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
axes[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
