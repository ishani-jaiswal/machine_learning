```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
iris = pd.read_csv('/content/drive/MyDrive/iris/iris.data')
```

```python
scaler = StandardScaler()
iris_scaled = scaler.fit_transform(iris.iloc[:, :-1])
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(iris_scaled)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(iris_scaled)
```

```python
plt.scatter(principalComponents[:, 0], principalComponents[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

```python
wine = pd.read_csv('/content/drive/MyDrive/winequality-red.csv',sep=";")
wine
```

```python
scaler = StandardScaler()
wine_scaled = scaler.fit_transform(wine.iloc[:, :-1])
```

```python
pca = PCA()
wine_pca = pca.fit_transform(wine_scaled)
```

```python
plt.figure(figsize=(10,5))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
n_components = 2
```

```python
pca_optimal = PCA(n_components=n_components)
wine_pca_optimal = pca_optimal.fit_transform(wine_scaled)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(wine_pca_optimal[:, 0], wine_pca_optimal[:, 1])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
